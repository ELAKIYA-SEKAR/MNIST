```python
import tensorflow
from tensorflow import keras
from tensorflow.keras.layers import Dense,Flatten
from tensorflow.keras import Sequential
```

```python
(x_train,y_train),(x_test,y_test)=keras.datasets.mnist.load_data()
```

```python
x_train_=x_train/255
x_test_=x_test/255
```

```python
x_train_[0]
```

```python
model=Sequential()
model.add(Flatten(input_shape=(28,28))) #converts into 1D=784 inputs
model.add(Dense(128,activation='relu'))
model.add(Dense(10,activation='softmax'))
```

```python
model.summary()
```

```python
model.compile(loss='sparse_categorical_crossentropy',optimizer='Adam',metrics=['Accuracy']) #in sparse it doesnt do one hot encoding but in cate crossentropy it does one hot encoding
```

```python
history=model.fit(x_train_,y_train,epochs=10, validation_split=0.2)
```

```python
y_prob=model.predict(x_test_)
```

```python
x_test_.shape
```

```python
x_test.shape
```

```python
y_prob.shape
```

```python
y_pred=y_prob.argmax(axis=1)
```

```python
y_pred.shape
```

```python
y_test.shape
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
```

```python
plt.plot(history.history['Accuracy'])
plt.plot(history.history['val_Accuracy'])
```
